# 3. Memory architectures from the proposal

This notebook demonstrates the five memory strategies used in the tutorial framework.

Architectures:
1. no memory
2. verbatim retrieval
3. extracted facts
4. episodic summaries
5. hybrid memory

## Proposal architecture taxonomy

| Architecture | Advantages | Disadvantages | Use Case |
|---|---|---|---|
| Verbatim Retrieval (RAG-based) | Preserves full context, minimal preprocessing | Low precision, high storage costs | Early exploration (<50 experiments) |
| Fact Extraction (Structured) | High precision, efficient storage | Information loss, extraction errors | Mature projects (50–500 experiments) |
| Episodic Compression (Summarisation) | Scalability, pattern emphasis | Detail loss, summarisation biases | Long runs (>500 experiments) |
| Hybrid Multi-Tier | Balanced precision/recall/latency | Implementation complexity | Production deployments |

Failure classes used throughout the notebooks:
- Retrieval failures: wrong information retrieved.
- Utilisation failures: retrieved information ignored.
- Storage failures: information loss during encoding.
- Temporal failures: outdated information persistence.
